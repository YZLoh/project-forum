```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('~/Desktop/Project_Forum/edmw_data_raw.csv')
df.head(20)
```

```python
df['Comment'].size
```

```python
#Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
#Remove punctuation using Regular Expression
#df['Comment'] = df['Comment'].str.replace(r'[^\w\s]', '') 
#df.head(20)
```

```python
#Remove white spaces
df['Thread Title'] = df['Thread Title'].str.replace(r'\r|\n|\t','')
df['Comment'] = df['Comment'].str.replace(r'\r|\n|\t',' ')
df.head(20)
```

```python
#Remove GAGT/PCWX app automatic signatures
df['Comment'].replace(r'Sent from .* using GAGT', '', inplace = True, regex=True) 
df['Comment'].replace(r'Posted from PCWX using .*', '', inplace = True, regex=True) 
# ^^ will not remove the entire phone model but as much as possible try 
df['Comment'].replace(r'Posted with .* via PCWX', '', inplace = True, regex=True)
```

```python
#Remove admin bot comments
df.drop(df[df['Comment Author']=='hwzadmin'].index, inplace=True) 
```

```python
#Change all letters to lowercase
df['Comment']  = df['Comment'].apply(lambda x: ' '.join([word.lower() for word in str(x).split()])) 
df.head(20)
```

```python
df['Comment'].replace('', np.nan, inplace = True) #replace empty strings with NaN
df.dropna(inplace = True) #Drop NaN rows (originally empty strings)
```

```python
#Remaining no. of comments
df['Comment'].size 
```

```python
df.to_csv('~/Desktop/Project_Forum/edmw_data_clean.csv',index = False)
```
